```python
import pandas as pd
```

```python
df = pd.read_csv(
    '../data/auto.csv',
    sep = ',',
    header=0,
    names=['ID', 'CarNumber', 'Make_n_model', 'Refund', 'Fines', 'History']
)
```

```python
df.set_index('ID', inplace=True)
df
```

```python
counter = df.count()
print(counter)
```

```python
df.drop_duplicates(subset=['CarNumber', 'Make_n_model', 'Fines'], keep='last', inplace=True)
counter_new = df.count()
print(counter_new)
```

```python
missing_val = df.isnull().sum()
print(missing_val)
threshold = len(df) - 500
df.dropna(axis=1, thresh=threshold, inplace=True)
print()
missing_val = df.isnull().sum()
print(missing_val)
print()
df['Refund'] = df['Refund'].ffill()
missing_val = df.isnull().sum()
print(missing_val)
print()
mean_val = df['Fines'].mean()
df['Fines'] = df['Fines'].fillna(mean_val)
missing_val = df.isnull().sum()
print(missing_val)
```

```python
df['Make'] = df.apply(lambda row: row['Make_n_model'].split()[0], axis=1)
df['Model'] = df.apply(lambda row: row['Make_n_model'].split()[1] if len(row['Make_n_model'].split()) > 1 else 'Witout model', axis=1)
df.drop(columns='Make_n_model', inplace=True) 
df
```

```python
df.to_json('../data/auto.json', orient='records')
```
