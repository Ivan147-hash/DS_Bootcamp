```python
import pandas as pd
import sqlite3
```

#### Подключение к базе данных

```python
try:
    conn = sqlite3.connect('../data/checking-logs.sqlite')
    print("Успешное подключение к базе данных.")
except sqlite3.Error as e:
    print(f"Ошибка при подключении к базе данных: {e}")
    exit()
```

#### Дата и время из таблицы pageviews только для пользователей

```python
query_views = "SELECT * FROM pageviews WHERE uid LIKE 'user_%'"
views = pd.read_sql_query(query_views, conn)
views
```

```python
df = pd.DataFrame(views)

# Преобразование столбца 'Дата' в тип datetime
df['datetime'] = pd.to_datetime(df['datetime'])

df['date'] = df['datetime'].dt.date

# Группировка по дате и подсчет суммы просмотров
df_viewe = df.groupby('date').size().reset_index(name='view')
df_viewe
```

#### Даты, когда были и просмотры, и коммиты checker

```python
query_commits = "SELECT * FROM checker WHERE uid LIKE 'user_%'"
commits = pd.read_sql_query(query_commits, conn)
commits
```

```python
df = pd.DataFrame(commits)

# Преобразование столбца 'Дата' в тип datetime
df['timestamp'] = pd.to_datetime(df['timestamp'])

df['date'] = df['timestamp'].dt.date

# Группировка по дате и подсчет суммы просмотров
df_commit = df.groupby('date').size().reset_index(name='commits')
df_commit
```

#### Объединение таблиц df_viewe и df_commit

```python
df_group = pd.merge(df_viewe, df_commit, on='date')
df_group
```

### Построение графика

```python
# Построение графика
ax = df_group.plot(x='date', y=['view', 'commits'], kind='line',
              figsize=(15,8), color=['r', 'c'], title='Views and commits per day',
              style=['-.', '--'], rot=90, fontsize=8)
ax.grid(axis='x')
ax.get_lines()[0].set_marker('o')
```

### Сколько раз количество просмотров превышало 150?

```python
count = df_group[df_group['view'] > 150].count()['view']
print(f'The answer is {count}')
```
