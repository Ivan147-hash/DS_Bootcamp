```python
import pandas as pd
import sqlite3
import matplotlib.pyplot as plt
import seaborn as sns
```

#### Подключение к базе данных

```python
df = pd.read_csv('../data/ab-test.csv')
df
```

#### Функция для отрисовки среднего значения каждой группы

```python
def add_mean_arrows(ax, data, group_labels):
    for i, group in enumerate(group_labels):  # Итерируемся по группам
        mean_value = data[data['group'] == group]['diff'].mean()
        x_pos = i + 1  # Позиция ящика на оси x (1 для первой группы, 2 для второй)
        ax.plot(x_pos, mean_value, marker='^', color='green', markersize=8) 
```

### Построение диаграммы

```python
fig, axes = plt.subplots(1, 2, figsize=(15, 8))

after_data = df[df['time'] == 'after']
before_data = df[df['time'] == 'before']

after_data.boxplot(column='diff', by='group',
                    ax=axes[0], color=dict(whiskers='red',
                    medians='red', caps='red'),
                    boxprops=dict(color = 'green', linewidth=3))
axes[0].set_title('After')
axes[0].set_xlabel('[group]')
add_mean_arrows(axes[0], after_data, after_data['group'].unique())

before_data.boxplot(column='diff', by='group',
                    ax=axes[1], color=dict(whiskers='red',
                    medians='red', caps='red'),
                    boxprops=dict(color = 'green', linewidth=3))
axes[1].set_title('Before')
axes[1].set_xlabel('[group]')
add_mean_arrows(axes[1], before_data, before_data['group'].unique())

plt.suptitle('Lag before labs deadlines. After and before the Newsfeed', fontsize=15)
plt.show()
```

### Каков был IQR контрольной группы до появления новостей?
До появления новостей (группа “before”) - IQR контрольной группы составляет примерно 80.

IQR — это межквартильный размах, равный Q3 - Q1.

Q1 (25-й процентиль) находится примерно в районе -140. Q3 (75-й процентиль) находится примерно в районе -60.

IQR = Q3 - Q1 = -60 - (-140) = 80.
