# Что можно сказать о фильмах?

##### Каждый год выпускалось множество фильмов. Но в какие годы было наибольшее количество релизов?

```python
dist_by_release = movies.dist_by_release()
print("Топ-10 лет по количеству выпущенных фильмов")
for i, (key, value) in enumerate(dist_by_release.items()):
    if i < 10:
        print(f'В {int(key)} г. было выпущено {value} фильмов')
    else:
        break
%timeit -n 10 -r 5  movies.dist_by_release()
```

##### Какие жанры самые популярные?

```python
dist_by_genres = movies.dist_by_genres()
dist_by_genres_proportion = movies.dist_by_genres_proportion()
for i, (key, value) in enumerate(dist_by_genres.items()):
    if i < 10:
        value_per = dist_by_genres_proportion[key]
        print(f'За все время было снято {value} в жанре {(key)}')
        print(f'---или {value_per}%')
    else:
        break
%timeit -n 10 -r 5 movies.dist_by_genres()
%timeit -n 10 -r 5 movies.dist_by_genres_proportion()
```

##### А какие фильмы объединяют в себе наибольшее количество жанров?

```python
dist_by_genres = movies.most_genres(10)
for i, (key, value) in enumerate(dist_by_genres.items()):
        print(f'Фильм {(key)} содержит {value} жанров')
%timeit -n 10 -r 5 movies.most_genres(10)
```

# Что можно сказать о рэйтинге фильмов?

##### Сколько оценок было оставлено в последние годы?

```python
dist_by_year = ratings.movies.dist_by_year()
for key, value in list(dist_by_year.items())[-10:]:
    print(f'В {int(key)} г. было поставлено {value} оценок')
%timeit -n 10 -r 5 ratings.movies.dist_by_year()
```

##### А как часто встречались разные оценки?

```python
dist_by_rating = ratings.movies.dist_by_rating()
for key, value in list(dist_by_rating.items()):
    print(f'Оценка {key} была использована за все время {value} раз')
%timeit -n 10 -r 5 ratings.movies.dist_by_rating()
```

##### Какие фильмы получили наибольшее количество оценок?

```python
top_by_num_of_ratings = ratings.movies.top_by_num_of_ratings(10)
for key, value in list(top_by_num_of_ratings.items()):
    print(f'Фильм {key} за все время получил {value} оценок')
%timeit -n 10 -r 5 ratings.movies.top_by_num_of_ratings(10)
```

##### Какие фильмы самые высокооцененные за все время?

```python
top_by_ratings_av = ratings.movies.top_by_ratings(10)
print("Топ-10 фильмов по среднему рейтингу")
for key, value in list(top_by_ratings_av.items()):
    print(f'Средняя оценка фильма {key} за все время - {value}')
%timeit -n 10 -r 5 ratings.movies.top_by_ratings(10)
```

```python
top_by_ratings_median = ratings.movies.top_by_ratings(10, 'median')
print("Топ-10 фильмов по медианному рейтингу")
for key, value in list(top_by_ratings_median.items()):
    print(f'Медианная оценка фильма {key} за все время - {value}')
%timeit -n 10 -r 5 ratings.movies.top_by_ratings(10, 'median')
```

##### Какие фильмы отличаются наибольшей дисперсией оценки?

```python
top_controversial = ratings.movies.top_controversial(10)
top_controversial_std_dev = ratings.movies.top_controversial_std_dev(10)
for key, value in list(top_controversial.items()):
    value_std = top_controversial_std_dev[key]
    print(f'Дисперсия оценки фильма {key} за все время - {value}')
    print(f"--стандартное отклонение - {value_std}")
%timeit -n 10 -r 5 ratings.movies.top_controversial(10)
%timeit -n 10 -r 5 ratings.movies.top_controversial_std_dev(10)
```

# Что можно сказть про анализ сервиса IMDb?

##### Какие режисеры сняли больше всего фильмов?

```python
top_10 = links.top_directors(10)
print('Топ-10 режисеров:')
for dir in top_10:
    print(dir, 'снял фильмов:', top_10[dir])
%timeit -n 10 -r 5 links.top_directors(3)
```

##### У каких фильмов самый большой бюджет?

```python
top_10 = links.most_expensive(10)
print('Топ-10 фильмов по сумме бюджета:')
for film in top_10:
    print('Бюджет фильма', film, 'составляет:', top_10[film])
%timeit -n 10 -r 5 links.most_expensive(3)
```

##### Какие фильмы получили больше всего прибыли после проката?

```python
top_10 = links.most_profitable(10)
print('Топ-10 фильмов по прибыли от проката:')
for profit in top_10:
    print('Фильм', profit, 'получил в прокате:', top_10[profit])
%timeit -n 10 -r 5 links.most_profitable(3)
```

##### Какие фильмы самые продолжительные?

```python
res = links.longest(10)
print('Топ-10 фильмов по продолжительности сеанса:')
for film in res:
    print('Продолжительность фильма', film, 'составляет:', res[film])
%timeit -n 10 -r 5 links.longest(3)
```

##### На какие фильмы потратили больше денег за каждую снятую минуту фильма?

```python
res = links.top_cost_per_minute(10)
print('Топ-10 фильмов по потраченным в минуту деньгам:')
for dir in res:
    print('На сьемку фильма', dir, 'потратили:', res[dir], '$/min')
%timeit -n 1 -r 3 links.top_cost_per_minute(3)
```

##### Какие фильмы принесли больше всего прибыли за каждую снятую минуту?

```python
res = links.top_cost_profit_minute(10)
print('Топ-10 фильмов по прибыли за каждую минуту:')
for dir in res:
    print('Фильм', dir, 'принес:', res[dir], '$ за каждую снятую минуту')
%timeit -n 1 -r 3 links.top_cost_profit_minute(3)
```
